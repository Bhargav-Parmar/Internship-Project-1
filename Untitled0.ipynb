Importing necessary libraries

```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

Trainng the model and working it out

```python
# Download data
df = yf.download("AAPL", period="2y", interval="1d")
df = df.reset_index()
df['Day_Count'] = np.arange(len(df))

# Train
X = df[['Day_Count']]
y = df['Close']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

model = LinearRegression()
model.fit(X_train, y_train)
```

Prediction and Visualization

```python
# Predict and Plot
predictions = model.predict(X_test)

plt.figure(figsize=(12, 6))
plt.scatter(df['Date'], df['Close'], color='blue', s=5, label='Actual Data')
plt.plot(df['Date'].iloc[len(X_train):], predictions, color='red', linewidth=2, label='Prediction Line')
plt.title('Stock Price Prediction: AAPL')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.show()
```
